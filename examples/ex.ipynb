```python
from sklearn.datasets import load_iris
import numpy as np
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt

from NNSOM import SOM
```

```python

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

X = np.transpose(X)
```

```python
# Set parameters
SOM_Row_Num = 4
Dimensions = (SOM_Row_Num, SOM_Row_Num)
Epochs = 200
Steps = 100
Init_neighborhood = 3
SEED = 1234567
```

```python
# Train SOM
# Initialize weights of SOM
som = SOM(Dimensions)
som.init_w(X)

# Train network
som.train(X, Init_neighborhood, Epochs, Steps)
```

```python
# Plot the hit histogram
fig4, ax4, patch4, text4 = som.hit_hist(X, True)
plt.show()
```

```python
# Plot the topology
fig0, ax0, patch0, text0 = som.plt_top_num()
plt.title('SOM Topology')
plt.show()
```

```python
# Plot distances between clusters
fig55, ax55, patch55 = som.neuron_dist_plot()
plt.title('SOM Neighbor Weight Distances', fontsize=16)
plt.show()
```
